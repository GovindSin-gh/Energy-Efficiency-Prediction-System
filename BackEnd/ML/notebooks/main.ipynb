### Data Preprocessing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('ENB2012_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.corr()
```

```python
import seaborn as sns
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True)
```

```python
## Independent and Dependent features
X=df.iloc[:,:-2] #independent features
y1=df.iloc[:,-2] #1st dependent variable 
y2=df.iloc[:,-1] #2nd dependent variable 
```

```python
X.head()
```

```python
y1
```

```python
y2
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y1_train, y1_test , y2_train , y2_test= train_test_split(X, y1, y2,test_size=0.25, random_state=42)
```

```python
X_train.shape,y1_train.shape,y2_train.shape,X_test.shape,y1_test.shape , y2_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

### Model Training for y1

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression
```

```python
regression.fit(X_train,y1_train)
```

```python
## slope or coefficients
regression.coef_
```

```python
## intercepts
regression.intercept_
```

```python
## Prediction for the test data
y1_pred=regression.predict(X_test)
```

```python
y1_test
```

```python
y1_pred
```

```python
## Performance Metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y1_test,y1_pred))
print(mean_absolute_error(y1_test,y1_pred))
print(np.sqrt(mean_squared_error(y1_test,y1_pred)))
```

```python
## R square and adjusted R square

from sklearn.metrics import r2_score
score=r2_score(y1_test,y1_pred)
print(score)
```

```python
## Adjusted R square
#display adjusted R-squared
1 - (1-score)*(len(y1_test)-1)/(len(y1_test)-X_test.shape[1]-1)
```

```python
from sklearn.model_selection import KFold , cross_val_score
```

```python
cv = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(regression,X, y1, cv=cv, scoring='r2')
print(f"CV R2 mean  std:{cv_scores.mean()}  {cv_scores.std()}")
```

```python
plt.scatter(y1_test,y1_pred)
plt.xlabel("Test Truth Data")
plt.ylabel("Test Predicted Data")
```

```python
residuals=y1_test-y1_pred
residuals
```

```python
sns.displot(residuals,kind="kde")
```

```python
## SCatter plot with predictions and residual
##uniform distribution
plt.scatter(y1_pred,residuals)
```

```python
from sklearn.linear_model import LassoCV
lassocv=LassoCV(cv=5)
lassocv.fit(X_train,y1_train)
y1_pred=lassocv.predict(X_test)
plt.scatter(y1_test,y1_pred)
mae=mean_absolute_error(y1_test,y1_pred)
score=r2_score(y1_test,y1_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
```

```python
lassocv.alpha_
```

```python
from sklearn.linear_model import RidgeCV
ridgecv=RidgeCV(cv=5)
ridgecv.fit(X_train,y1_train)
y1_pred=ridgecv.predict(X_test)
plt.scatter(y1_test,y1_pred)
mae=mean_absolute_error(y1_test,y1_pred)
score=r2_score(y1_test,y1_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
```

```python
ridgecv.alpha_
```

```python
from sklearn.linear_model import ElasticNetCV
elasticcv=ElasticNetCV(cv=5)
elasticcv.fit(X_train,y1_train)
y1_pred=elasticcv.predict(X_test)
plt.scatter(y1_test,y1_pred)
mae=mean_absolute_error(y1_test,y1_pred)
score=r2_score(y1_test,y1_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
y1_ridge=Ridge(alpha=0.1)
y1_ridge.fit(X_train,y1_train)
y1_pred=y1_ridge.predict(X_test)
mae=mean_absolute_error(y1_test,y1_pred)
score=r2_score(y1_test,y1_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter(y1_test,y1_pred)
```
